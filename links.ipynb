```python
from sage.plot.plot3d.shapes import *
from sage.plot.plot3d.shapes2 import Line
```

```python
def palette(col, n):
    if col == 'red':
        return (1 - n/8,0.2 + n/10,0)
    elif col == 'blue':
        return (0,0.5 + n/10,1 - n/8)

Faces =  Sphere(.1, color='red',opacity = 0)
Faces += Sphere(.1, color='blue', opacity = 0).translate(0,.01,0)
    
w = [9,4,6]
x = [9,13.5,6]
y = [-9,13.5,6]
z = [-9,4,6]
ww = [9,4,-6]
xx = [9,13.5,-6]

Face_1 = polygon3d([w,x,y,z], alpha = 1, color='lightblue')
Face_2 = polygon3d([w,x,xx,ww], alpha = 1, color='lightgray')
Face_3 = Face_1 + Face_2
Face_4 = Face_3 + Face_3.rotate((0,1,0), math.pi)
Face_5 = Face_4 + Face_4.rotate((0,0,1), math.pi)
Faces += Face_5


OuterEdges = Sphere(1, opacity=0)


for i in range(4):
    Lines_1 = LineSegment(w,z, 10, color='blue').rotate((0,1,0),i * math.pi)
    Lines_2 = LineSegment(w,ww, 10, color='blue').rotate((0,1,0),i * math.pi)
    Lines_3 = Lines_1 + Lines_2
    Lines_4 = Lines_3 + Lines_3.rotate((0,0,1),math.pi)
    OuterEdges += Lines_4
#     OuterEdges += LineSegment([1,-2,1], [-1,-2,1], 4, color=palette('red',i)).rotate((0,1,0),i * math.pi/2)

    
# InnerEdges = OuterEdges.rotate((1,0,0),math.pi/2).scale(0.5,0.5,0.5)

Quotient = Faces + OuterEdges
Quotient.plot()
```

```python
nested = [Quotient]



for i in range(8):
    nested += [nested[-1].scale(2/3,2/3,2/3).rotate((1,0,0),math.pi/2).rotate((0,1,0),math.pi/2)]

sum(nested).plot()
```
