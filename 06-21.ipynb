```python
from flatsurf.geometry.similarity_surface_generators import HalfDilationSurfaceGenerators
s = HalfDilationSurfaceGenerators.test_triangle(1/2, 1/5, 1/7)
show(s.plot())

sc = s.canonicalize()
show(sc.plot())
print(sc.num_polygons())
scc = sc.canonicalize()
show(scc.plot())
print(sc.num_polygons())
```
