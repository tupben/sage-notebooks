```python
# With a twist:
# 
# The gluings of this base surface generate a cover whose nested chambers
# 'twist' 90 degrees through the x,y and z axes at each layer.
```

```python
from sage.plot.plot3d.shapes import *
from sage.plot.plot3d.shapes2 import Line
```

```python
def palette(col, n):
    if col == 'red':
        return (1 - n/8,0.2 + n/10,0)
    elif col == 'blue':
        return (0,0.5 + n/10,1 - n/8)
```

```python
# b is the ratio of the side lengths within a rectangle (z/x).
# c is the ratio or the entire chamber length to the width of a rectangle (y/x),
# c is the linear scaling factor between each nested chamber.

b = 1
c = 2

Faces =  Sphere(.1, color='red',opacity = 0)
Faces += Sphere(.1, color='blue', opacity = 0).translate(0,.01,0)
   
    
Face_1 = polygon3d([[1/2,c/2,b/2], [-1/2,c/2,b/2],[-1/2,b/(2*c),b/2],[1/2,b/(2*c),b/2]], alpha = 0.9, color='lightgray')
Face_2 = polygon3d([[1/2,b/(2*c),b/2],[b*b/(2*c),b/(2*c),b/2],[b*b/(2*c),-1*b/(2*c),b/2],[1/2,-1*b/(2*c),b/2]], alpha = 0.9, color='lightgray')
Face_3 = Face_1 + Face_2
Face_4 = Face_3 + Face_3.rotate((0,0,1),math.pi)
Face_5 = Face_4 + Face_4.rotate((1,0,0),math.pi)
Face_6 = polygon3d([[1/2,c/2,b/2],[1/2,c/2,-1*b/2],[1/2,-1*c/2,-1*b/2],[1/2,-1*c/2,b/2]], alpha = 0.9, color='lightgray')
Face_7 = Face_6 + Face_6.rotate((0,1,0),math.pi)
Faces += Face_5 + Face_7

OuterEdges = Sphere(1, opacity=0)

for i in range(2):
    OuterEdges += LineSegment([1/2,c/2,b/2], [-1/2,c/2,b/2], 1, color=palette('blue',2*i)).rotate((0,1,0),i * math.pi)
    OuterEdges += LineSegment([-1/2,c/2,b/2], [-1/2,c/2,-b/2], 1, color=palette('blue',2*i + 1)).rotate((0,1,0),i * math.pi)
    OuterEdges += LineSegment([1/2,-1*c/2,b/2], [-1/2,-1*c/2,b/2], 1, color=palette('red',2*i)).rotate((0,1,0),i * math.pi)
    OuterEdges += LineSegment([-1/2,-1*c/2,b/2],[-1/2,-1*c/2,-1*b/2], 1, color=palette('red',2*i+1)).rotate((0,1,0),i * math.pi)

    
InnerEdges = OuterEdges.rotate((1,0,0),1*math.pi/2).rotate((0,0,1),1*math.pi/2).scale(b/c,b/c,b/c)

Quotient = Faces + InnerEdges + OuterEdges
Quotient.plot()
```

```python
nested = [Quotient]

for i in range(10):
    nested += [nested[-1].scale(b/c,b/c,b/c).rotate((1,0,0),1*math.pi/2).rotate((0,0,1),1*math.pi/2)]

sum(nested).plot()
```
