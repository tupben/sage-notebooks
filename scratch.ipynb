```python
s = Surface_list(QQ)
rect = polygons(vertices = ((0,0),(1,0),(1,1),(0,1)))
s.add_polygon(rect)
s.set_edge_pairing(0,0,0,2)
s.set_edge_pairing(0,1,0,3)

S = TranslationSurface(s)

S.plot()

DS = S.minimal_cover(cover_type='dilation').copy(relabel=True)
IDC1 = IDC(DS, 500, veech = 1, show_current_sym = 0) # Change 1 to > 100
show(IDC1.plot(model = 'UHP', figsize = 10)) #UHP or KM
show(list(IDC1.generators(normalize_det=1)))
```
