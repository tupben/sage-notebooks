```python
# Example of a TRANSLATION SURFACE.

# Define a surface. Whta is QQ? The field of rational numbers.
s = Surface_list(QQ)

# Define a polygon.
square = flatsurf.polygons.regular_ngon(4)

# Add the polygon to the surface.
s.add_polygon(square)
    
# Add edge gluings.
s.change_edge_gluing(0,0,0,1)
s.change_edge_gluing(0,2,0,3)

# Make into a cone surface
cs = ConeSurface(s)
# cs.plot()

# Make into a translation surface.
ts = cs.minimal_cover(cover_type="translation")
# ts.plot(polygon_labels=False, edge_labels=False)
ts.plot()
```

```python
# This one is a SIMILARITY SURFACE.

# Define a surface.
s = Surface_list(QQ)

# Define a polygon.
P = polygons(vertices=[(0,0),(1,0),(1,1),(0,1)])

# Add the polygon to the surface (several times).
s.add_polygon(P)
s.add_polygon(2*P)
s.add_polygon(3*P)

# Pair edges.
s.set_edge_pairing(0,1,1,3)
s.set_edge_pairing(0,0,2,2)
s.set_edge_pairing(0,2,2,0)
s.set_edge_pairing(0,3,1,1)
s.set_edge_pairing(1,2,2,1)
s.set_edge_pairing(1,0,2,3)

# Make a similarity surface.
S = SimilaritySurface(s)
S.plot()
```

```python
s = Surface_list(QQ)
P = flatsurf.polygons(vertices = [(0,0),(3,0),(3,3),(2,4),(1,4),(0,3)])

s.add_polygon(P)
s.add_polygon(P)
s.add_polygon(P)
s.add_polygon(P)

S = SimilaritySurface(s)

for i in range(4):
    s.set_edge_pairing(i,0,i,3)
    s.set_edge_pairing(i,1,(i+1)%4,5)
    s.set_edge_pairing(i,2,(i+1)%4,4)

gs = S.graphical_surface()
gs.make_adjacent(0,2)
gs.make_adjacent(1,2)
gs.plot()
```
