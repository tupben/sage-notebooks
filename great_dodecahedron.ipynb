```python
from sage.plot.plot3d.shapes import *
from sage.plot.plot3d.shapes2 import Line
import math
```

```python
# Here is a SAGE graph.

def palette(col, n):
    if col == 'red':
        return (1 - n/8,0.2 + n/10,0)
    elif col == 'blue':
        return (0,0.5 + n/10,1 - n/8)


OuterEdges = Sphere(1, opacity=0)
Origin =  Sphere(0.1, color='black',opacity = 1)
Origin += Sphere(.1, color='white', opacity = 1).translate(0,.005,0)    

Star = polygon3d([[0,-1,0], [1,-1,0],[1,-1,1]], alpha = 0.0, color='lightgray')

u = 1/(2*math.tan(math.pi/5))
v = math.acos(-1/math.sqrt(5))
w = (math.tan(math.pi/5))/2
golden = (1 + 5 ** 0.5) / 2

Fill_1 = polygon3d([[1/2,0,0],[-1/2,0,0],[0,w+u,0]], alpha = 0.9, color='lightgray').translate(0,u,0)

Line_1 = LineSegment([1/2 + golden,u,0],[-1 * (1/2 + golden),u,0], 1, color = "blue")
Line_2 = LineSegment([1/2 + golden,u,0],[-1 * (1/2 + golden),u,0], 1, color = "red")

Star_1 = Sphere(0.1, color='black',opacity = 0)
Star_2 = Sphere(0.1, color='black',opacity = 0)

for i in range(5):
    Star_1 += Fill_1.rotate((0,0,1),2 * i * math.pi / 5).translate(u,0,0)
#     Star_1 += Line_1.rotate((0,0,1),2 * i * math.pi / 5).translate(0,-1*u,0)
#     Star_2 += Line_2.rotate((0,0,1),2 * i * math.pi / 5).translate(0,-1*u,0).rotate((1,0,0),v)
    Star_2 += Fill_1.rotate((0,0,1),2 * i * math.pi / 5).translate(0,0,0).rotate((0,1,0),math.pi/5)
    
Chair = Sphere(0.1, color='black',opacity = 0)
for j in range(5):
    Chair += (Star_1 + Star_2).translate(0,u,0).rotate((0,0,1), j * 2 * math.pi / 5)
#     Chair.translate(100,u,0)
    
Total = Fill_1 + Origin
Total.plot()
```
