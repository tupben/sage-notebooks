```python
import flatsurf

# Choose translation surface and trajectory
T = flatsurf.polygons.triangle(9,6,3)
B = flatsurf.similarity_surfaces.billiard(T)
v = B.tangent_vector(0,(0,0.5), (1,3))
l = v.straight_line_trajectory()

# Label edges
alphabet = {(0,0):'A',(0,1):'B',(0,2):'C'}

l.flow(100);

# Run
print(''.join(l.coding(alphabet)))
l.plot()
```
