# Affine Symmetries

```python
s = TranslationSurfaceGenerators.square_torus()
```

```python
s.plot()
```

```python
M = matrix([[1, 3/2],
            [1,-1]])
M
```

```python
s2 = M*s
s2.plot()
```

```python
M = matrix([[1, 1],
            [0, 1]])
M
s2 = M*s
show(s.plot())
s2.plot()
```

```python
s2c = s2.canonicalize()
s2c.plot()
```

```python
s.canonicalize() == (M*s).canonicalize()
```

```python
s = TranslationSurfaceGenerators.regular_octagon()
```

```python
s.plot()
```

```python
R = s.base_ring()
R
```

```python
a = R.gen()
```

```python
a.n()
```

```python
M = matrix([
    [1, 2+2*a],
    [0, 1]
])
```

```python
s.canonicalize().plot()
```

```python
s2 = M*s
s2.plot()
```

```python
s2.canonicalize().plot()
```

```python
s.canonicalize() == (M*s).canonicalize()
```

```python
R = matrix([
    [a/2, -a/2],
    [a/2,  a/2],
])
```

```python
R.n()
```

```python
s.canonicalize() == (R*s).canonicalize()
```

```python
N = M*R*M*~R*~R
```

```python
s.canonicalize() == (N*s).canonicalize()
```
