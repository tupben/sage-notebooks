```python
from sage.plot.line import Line

def grid(a,b):
    lines = line([(a,0), (b,0)], color = 'red', thickness=1)
    lines += line([(0,a), (0,b)], color = 'red', thickness=1)
    for i in range(a,b+1):
        lines += line([(a,i), (b,i)], color = 'gray', thickness=1)
        lines += line([(i,a), (i,b)], color = 'gray', thickness=1)
    return lines
```

# Affine Symmetries

```python
s = TranslationSurfaceGenerators.square_torus()
```

```python
s.plot()
```

```python
M = matrix([[1, 3/2],
            [1,-1]])
M
```

```python
s2 = M*s
s2.plot()
```

```python
M = matrix([[1, 1],
            [0, 1]])
M
s2 = M*s
s2.plot()
```

```python
s2c = s2.canonicalize()
s2c.plot()
```

```python
s.canonicalize() == (M*s).canonicalize()
```

```python
s = TranslationSurfaceGenerators.regular_octagon()
```

```python
s.plot()
```

```python
R = s.base_ring()
R
```

```python
a = R.gen()
```

```python
a.n()
```

```python
M = matrix([
    [1, 2+2*a],
    [0, 1]
])
```

```python
s2 = M*s
s2.plot()
```

```python
s2.canonicalize().plot()
```

```python
s.canonicalize() == (M*s).canonicalize()
```

```python
N = matrix([
    [1, 1],
    [0, 1]
])
N
```

```python
t = N*s
# t.plot() + t.canonicalize().plot() + grid(-2,4)
t.plot()  + grid(-2,4)
```

```python
t.canonicalize().plot() + grid(-2,4)
```

```python
t.canonicalize() == s.canonicalize()
```
